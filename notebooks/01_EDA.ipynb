```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import duckdb

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
PARQUET_PATH = '../data/processed/Electric_Vehicle_Population_Data.parquet'
df = pd.read_parquet(PARQUET_PATH)

print(f"Dataset Shape: {df.shape}")
df.info()
```

```python
print("Summary Statistics:")
display(df.describe())
```

```python
print("Top 15 Counties by EV Registration:")
county_counts = df['County'].value_counts().head(15)
sns.barplot(x=county_counts.values, y=county_counts.index, hue=county_counts.index, palette="viridis")
plt.title("Top 15 Counties by EV Registration")
plt.show()
```

```python
print("EV Adoption Trend by Model Year:")
adoption_trend = df['Model Year'].value_counts().sort_index()
sns.lineplot(x=adoption_trend.index, y=adoption_trend.values, marker='o', color='purple')
plt.title("EV Adoption Trend by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles")
plt.show()
```

```python
print("BEV vs PHEV Distribution:")
type_counts = df['Electric Vehicle Type'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['#0891b2', '#a855f7'])
plt.title("BEV vs PHEV Distribution")
plt.show()
```

```python
print("CAFV Eligibility Breakdown:")
cafv_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
sns.barplot(x=cafv_counts.values, y=cafv_counts.index, hue=cafv_counts.index, palette="magma")
plt.title("CAFV Eligibility Breakdown")
plt.show()
```
